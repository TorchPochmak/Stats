```python
import math

fig, ax = plt.subplots(figsize=(5, 5))
ax.set(xlim=(-6, 8), ylim=(-30, 50))

x = list(range(-5, 7))
y = np.array([i**2 for i in x])
y1 = np.array([i**2+5 for i in x])

arrows = ['-', '->', '-[', '|-|', '-|>', '<-', '<->', '<|-', '<|-|>',
'fancy', 'simple', 'wedge']
for i in range(len(x)):
    plt.annotate(str(i), xy=(x[i], y[i]), xycoords='data', xytext=(x[i]+0.1, y[i]-10),
    textcoords='data', arrowprops=dict(facecolor='black', arrowstyle=arrows[i]))
for i in range(len(x)):
    plt.plot([x[i], x[i]], [y[i], y1[i]], 'black')

plt.plot(x, y)
plt.plot(x, y1)
plt.plot(x, y, 'ro')
plt.plot(x, y1, 'ro')
```
