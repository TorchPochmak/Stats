```python
# xlabel (или ylabel): 
# str - текст
# labelpad: расстояние до графика

# выравнивание loc=

#Принимают параметры конструктора класса matplotlib.text.Text в качестве args
# fontsize или size, fontweight, color, fontstyle и т.д.
plt.xlabel('Day', labelpad=50, fontsize=15, color='blue', 
    fontweight='semibold', fontstyle='italic', loc='left')
```

```python
plt.title('YO', pad=50)
#pad - расстояние
#loc=, matplotlib.text.Text...
```

```python
#Делать подписи на графике - plt.text()
x = [1, 5, 10, 15, 20]
y = [1, 7, 3, 5, 11]
plt.plot(x, y, label='steel price')
plt.title('Chart price', fontsize=15)
plt.xlabel('Day', fontsize=12, color='blue')
plt.ylabel('Price', fontsize=12, color='blue')
plt.legend()
plt.grid(True)
plt.text(15, 4, 'grow up!')
```

```python
x = np.arange(0, 1.01, 0.01)
lines = plt.plot(x, np.sin(2*np.pi*x), x, np.sin(4*np.pi*x))

plt.setp(lines, linewidth=2, color='r') # редактирование части lines из plot
```

```python
#types of lines
x = [1, 5, 10, 15, 20]
y1 = [1, 7, 3, 5, 11]
y2 = [i*1.2 + 1 for i in y1]
y3 = [i*1.2 + 1 for i in y2]
y4 = [i*1.2 + 1 for i in y3]
plt.plot(x, y1, '-', x, y2, '--', x, y3, '-.', x, y4, ':')
```

```python
x = [1, 5, 10, 15, 20]
y1 = [1, 7, 3, 5, 11]
y2 = [i**1.2 for i in y1]
y3 = [i*1.2 + 1 for i in y2]
y4 = [i*1.2 + 1 for i in y3]
plt.plot(x, y1, 'ro',x, y2, 'bx', x, y3, 'b*', x, y4, 'g^')
```

```python
#subplot...
x = [1, 5, 10, 15, 20]
y1 = [1, 7, 3, 5, 11]
y2 = [i**1.2 for i in y1]
y3 = [i*1.2 + 1 for i in y2]
y4 = [i*1.2 + 1 for i in y3]

plt.figure(figsize=(6,6))
plt.subplot(221) #сокращенный вар от 2,2,1
plt.plot(x, y1, 'ro')
plt.subplot(222)
plt.plot(x, y2, 'bx')
plt.subplot(223)
plt.plot(x, y3, 'b*-')
plt.subplot(224)
plt.plot(x, y4, 'g^:')

#второй вариант, заранее задать размер сетки, не позволит по-смешному объединять
fig, axs = plt.subplots(2, 2, figsize=(12, 7))
axs[0, 0].plot(x, y1, '-')
axs[0, 1].plot(x, y2, '--')
axs[1, 0].plot(x, y3, '-.')
axs[1, 1].plot(x, y4, ':')
```
