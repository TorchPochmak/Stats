```python
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
np.random.seed(123)
vals = np.random.randint(100, size=(7, 7))
fig, ax = plt.subplots()
gr = ax.pcolor(vals)
axins = inset_axes(ax, width="7%", height="50%", loc='lower left',
bbox_to_anchor=(1.2, 0., 1, 1), bbox_transform=ax.transAxes,
borderpad=0)
#можно модифицировать шкалу с помощью объекта класса Tick или параметра ticks
plt.colorbar(gr, cax=axins, ticks=[i for i in range(0, 100, 10)], label='Value')
```
