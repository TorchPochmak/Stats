```python
import numpy as np
import scipy as sp
import seaborn as sns 
import matplotlib.pyplot as plt
import scipy.stats as sps
import pandas as pd
import ipywidgets as widgets

import random
import math

import itertools

%matplotlib inline
```

```python
x = np.linspace(0, 10, 100) #[0, 10] разбито на 100 частей
y1 = x
y2 = [i**2 for i in x]
y3 = [math.sin(i)*10 for i in x]

plt.title('plots', fontsize=16)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.plot(x, y1, 'b--', x, y2, 'r-.', x, y3, 'g')
```

```python
x = np.linspace(0, 10, 100) #[0, 10] разбито на 50 частей
y1 = x
y2 = [i**2 for i in x]

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
#первый - кол-во строк, второй - кол-во столбцов, третий - индекс (с 1)
plt.title('y = x')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.plot(x, y1, 'b--')

plt.subplot(1,2,2)
plt.title('y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.plot(x, y2, 'r-.')
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
AutoMinorLocator)
import numpy as np


x = np.linspace(0, 10, 11)
y1 = 4*x
y2 = [i**2 for i in x]

fig, ax = plt.subplots(figsize=(8, 8)) # тип ax - matplotlib.axes._axes.Axes

ax.set_title('Графики зависимостей: y1=4*x, y2=x^2', fontsize=16)

ax.set_xlabel('x', fontsize=14)
ax.set_ylabel('y1, y2', fontsize=14)

ax.grid(which='major', linewidth=1.2)
ax.grid(which='minor', linestyle='--', color='gray', linewidth=0.5)

ax.scatter(x, y1, c='red', label='y1 = 4*x') #натыкать точек на графике
ax.plot(x, y2, label='y2 = x^2')
ax.legend() #крутая штучка в углу

ax.xaxis.set_minor_locator(AutoMinorLocator()) #не подписывать, но разбивать
ax.yaxis.set_minor_locator(AutoMinorLocator())

ax.tick_params(which='major', length=5, width=2) #жирность черточек справа и снизу
ax.tick_params(which='minor', length=5, width=1)

plt.show()
```
